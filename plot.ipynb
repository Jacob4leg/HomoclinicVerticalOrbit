```python
curve = np.loadtxt("homoclinic.txt")

X = curve[:,0]
Y = curve[:,1]
Z = curve[:,2]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_zticklabels([])

ax.plot(X, Y, Z, color='b', linewidth=2)

# L = max(X.max(), Y.max(), Z.max(), -X.min(), -Y.min(), -Z.min())
L_X = X.max() + 0.2
L_Y = Y.max()
L_Z = Z.max() + 0.02
ax.plot([-L_X, L_X], [0, 0], [0, 0], color='k')
ax.plot([0, 0], [-L_Y, L_Y], [0, 0], color='k')
ax.plot([0, 0], [0, 0], [-L_Z, L_Z], color='k')

ax.scatter3D(muSJ,0,0, color='r')
ax.scatter3D(1 - muSJ,0,0, color='r')

ax.text(L_X, 0, 0, 'X', color='r')
ax.text(0, L_Y, 0, 'Y', color='g')
ax.text(0, 0, L_Z, 'Z', color='b')


ax.grid(True)
ax.set_box_aspect([1,1,1])

plt.savefig("homoclinic_orbit.pdf", bbox_inches='tight', pad_inches=0)

plt.show()
```
